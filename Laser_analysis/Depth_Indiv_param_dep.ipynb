```python
fig = plt.figure(figsize=(7,6))

# model1 = Pipeline([
#     ('linear_regression', LinearRegression())
# ])

grouped_df = AllExp.groupby(['Frequency', 'Speed', \
                                'Passes', 'Focal_Position'], as_index=False)
colors = ["g","r", "b", "k", "orange", "c", "m", 'brown', 'k']
markers = ["o", "*", "s", "p", "X", "P", "v", "h", "D", "d",  "^", "<", ">"]
LR2 = LinearRegression()
i = 0
for key, item in grouped_df:
    
    if item.shape[0] >= 5 :
        x = grouped_df.get_group(key).Amplitude
        y = grouped_df.get_group(key).Depth
        if max(x.values) > 0.1:
#             Model1 = model1.fit((np.array(x)).reshape(-1,1), y)
#             y_ln = Model1.predict((np.array(x)).reshape(-1,1))
            X2 = np.c_[np.ones(len(y)),x.values**(-1),x.values**(1),x.values**(2)\
                       ]
            result = LR2.fit(X2, y)
            xx = np.linspace(min(x.values), max(x.values), 30)
            Xcont = np.c_[np.ones(len(xx)), xx**(-1),xx,xx**(2)]
            yp = LR2.predict(Xcont)
            plt.plot(x,y, marker=markers[i%13],c = colors[i%9],lw=0, ms=10)
            plt.plot(xx,yp, '-',c = colors[i%9],ms=10)
#             plt.plot(x,y_ln, '-',c = colors[i%9],ms=10)
            i += 1

plt.text(0.1,600,'(a)',fontsize= 35)
plt.xlabel('Amplitude ($mm$)', fontsize=25)
plt.ylabel('Depth ($\\mu m$)', fontsize=25)
plt.tick_params(axis='x', labelsize=20)
plt.tick_params(axis='y', labelsize=20)
plt.savefig('Graphs/Depth_Amp.png',bbox_inches='tight')

plt.show()
```

```python
grouped_df = AllExp.groupby(['Speed', 'Amplitude', \
                                'Passes', 'Focal_Position'], as_index=False)


model1 = Pipeline([
    ('linear_regression', LinearRegression())
])
colors = ["g","r", "b", "k", "orange", "c", "m", 'brown']
markers = ["o", "*", "s", "p", "x", "X", "P", "v", "h", "D", "d",  "^", "<", ">"]

i = 0
for key, item in grouped_df:
    
    if item.shape[0] >= 0 :
        x = grouped_df.get_group(key).Frequency
        y = grouped_df.get_group(key).Depth
        Model1 = model1.fit((np.array(x)).reshape(-1,1), y)
        y_ln = Model1.predict((np.array(x)).reshape(-1,1))
        plt.plot(x,y, marker=markers[i%14],c = colors[i%8],lw=0, ms=10)
        plt.plot(x,y_ln, '-',c = colors[i%8],ms=10)
        i += 1


plt.xlabel('Frequency ($Hz$)', fontsize=25)
plt.ylabel('Depth ($\\mu m$)', fontsize=25)
plt.tick_params(axis='x', labelsize=20)
plt.tick_params(axis='y', labelsize=20)

plt.show()
```

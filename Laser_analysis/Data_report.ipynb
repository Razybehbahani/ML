```python
"""import packages you need"""
import numpy as np
import pandas as pd
```

```python
df_0 = pd.read_csv('../Data/Param0_2width.csv')
df_0.Passes = pd.to_numeric(df_0.Passes, downcast='integer')
df_0.Frequency = df_0.Frequency.round()
df_0.Frequency = pd.to_numeric(df_0.Frequency.values, downcast='integer')

rows=[]
for i in range(1,28):
    rows.append(i)
df_00 = df_0.set_index([rows])

df_00.to_csv(r'../Data/Tune_Exp0.csv', header=True, index = True)

df_00
```

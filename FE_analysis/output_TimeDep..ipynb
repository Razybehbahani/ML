```python
"""import packages you need"""
import numpy as np
import pandas as pd
import math
import os
import matplotlib.pyplot as plt 

from os import listdir
from os.path import isfile, join
```

```python


def plotgraph(File_address, title):
    edgeAngle = [f for f in listdir(File_address) if f.endswith('deg')]
    
    fig, axes = plt.subplots(dpi = 120,nrows = 3, ncols = 3,figsize=(12, 11))
    plt.subplots_adjust(hspace = 0.28, wspace=0.35)  #determines the distance between two rows of plots
    ax = axes.ravel()
        
    for i in range(len(edgeAngle)):
        
        textfile = [ file for file in os.listdir(File_address+edgeAngle[i]) if file.endswith('.csv')]
        for j in textfile:
            if (j == 'Safety factor.csv'):
                d = pd.read_csv(File_address +edgeAngle[i] + '/' + j, skiprows=5)
                ax[0].plot(d.iloc[:,1], label=edgeAngle[i])
                ax[0].set_ylabel('Safety_factor', fontsize=20) 
                ax[0].legend(loc = 1, fontsize=10)
                ax[0].set_xlabel('time', fontsize=15)
                bbox_propsLAD = dict(boxstyle="round4,pad=0.3", fc="white", ec="b", lw=1.5)
                ax[0].annotate('min', xy=(100, 8), fontsize=15,bbox=bbox_propsLAD,color = 'k')
    #             Safety_factor = d.iloc[:,1].min()


            elif (j == 'Different energies.csv'):
                d = pd.read_csv(File_address +edgeAngle[i]+ '/' + j, skiprows=5)
    #             Total_contact_energy =  d.iloc[:,1].sum()
    #             Elastic_strain_energy = d.iloc[:,3].max()
                ax[2].plot(d.iloc[:,1])
                ax[2].set_ylabel('Tot. Contact Engy', fontsize=20)
                ax[2].set_xlabel('time', fontsize=15)
                bbox_propsLAD = dict(boxstyle="round4,pad=0.3", fc="white", ec="b", lw=1.5)
                ax[2].annotate('max', xy=(200, 0.2), fontsize=15,bbox=bbox_propsLAD,color = 'k')
#                 ax[2].legend(loc = 1, fontsize=10)
                ax[3].plot(d.iloc[:,3])
                ax[3].set_ylabel('Elast. Strain Engy', fontsize=20)
                ax[3].set_xlabel('time', fontsize=15)
                bbox_propsLAD = dict(boxstyle="round4,pad=0.3", fc="white", ec="b", lw=1.5)
                ax[3].annotate('max', xy=(200, 5), fontsize=15,bbox=bbox_propsLAD,color = 'k')
#                 ax[3].legend(loc = 1, fontsize=10)

                
            elif (j == 'Maximum out-of-plane deformation.csv'):
                d = pd.read_csv(File_address +edgeAngle[i]+ '/' + j, skiprows=5)
    #             Max_oop_deformation = d.iloc[:,1].max()
                ax[1].plot(d.iloc[:,1])
                ax[1].set_ylabel('Oop Deform', fontsize=20)
                ax[1].set_title(title, fontsize=25)
                ax[1].set_xlabel('time', fontsize=15)
                bbox_propsLAD = dict(boxstyle="round4,pad=0.3", fc="white", ec="b", lw=1.5)
                ax[1].annotate('max', xy=(400, 0.2), fontsize=15,bbox=bbox_propsLAD,color = 'k')
                
            elif (j == 'Maximum temperature of the edge.csv'):
                d = pd.read_csv(File_address +edgeAngle[i]+ '/' + j, skiprows=5)
    #             Max_Edge_temp = d.iloc[:,1].max()
                ax[4].plot(d.iloc[:,1])
                ax[4].set_ylabel('Edge Temp.', fontsize=20)
                ax[4].set_xlabel('time', fontsize=15)
                bbox_propsLAD = dict(boxstyle="round4,pad=0.3", fc="white", ec="b", lw=1.5)
                ax[4].annotate('max', xy=(400, 150), fontsize=15,bbox=bbox_propsLAD,color = 'k')
            
            elif (j == 'Average friction force.csv'):
                d = pd.read_csv(File_address +edgeAngle[i]+ '/' + j, skiprows=5)
    #             Max_Ave_fric_force = d.iloc[:,1].mean()
                ax[5].plot(d.iloc[:,1])
                ax[5].set_ylabel('Avr. frictForce', fontsize=20)
                ax[5].set_xlabel('time', fontsize=15)
                bbox_propsLAD = dict(boxstyle="round4,pad=0.3", fc="white", ec="b", lw=1.5)
                ax[5].annotate('max', xy=(400, 10000), fontsize=15,bbox=bbox_propsLAD,color = 'k')

            elif (j == 'Total net heat rate.csv'):
                d = pd.read_csv(File_address +edgeAngle[i]+ '/' + j, skiprows=5)
    #             Max_heat_rate = d.iloc[:,1].max()
                ax[6].plot(d.iloc[:,1])
                ax[6].set_ylabel('Heat Rate', fontsize=20)
                ax[6].set_xlabel('time', fontsize=15)
                bbox_propsLAD = dict(boxstyle="round4,pad=0.3", fc="white", ec="b", lw=1.5)
                ax[6].annotate('max', xy=(400, 60), fontsize=15,bbox=bbox_propsLAD,color = 'k')
                

            elif (j == 'Internal energy.csv'):
                d = pd.read_csv(File_address +edgeAngle[i]+ '/' + j, skiprows=5)
    #             Max_internal_energy = d.iloc[:,1].max()
                ax[7].plot(d.iloc[:,1])
                ax[7].set_ylabel('Int.Engy', fontsize=20)
                ax[7].set_xlabel('time', fontsize=15)
                bbox_propsLAD = dict(boxstyle="round4,pad=0.3", fc="white", ec="b", lw=1.5)
                ax[7].annotate('max', xy=(400, 750), fontsize=15,bbox=bbox_propsLAD,color = 'k')

            elif (j == 'Friction dissipation rate.csv'):
                d = pd.read_csv(File_address +edgeAngle[i]+ '/' + j, skiprows=5)
    #             Total_friction_dissipation = d.iloc[:,1].sum() 
                ax[8].plot(d.iloc[:,1])
                ax[8].set_ylabel('FricDissipRate', fontsize=20)
                ax[8].set_xlabel('time', fontsize=15)
                bbox_propsLAD = dict(boxstyle="round4,pad=0.3", fc="white", ec="b", lw=1.5)
                ax[8].annotate('sum', xy=(400, 0.06), fontsize=15,bbox=bbox_propsLAD,color = 'k')

plt.show()
```

```python

def plot_graph(File_address,ax, edgeAngle, output, title):
    textfile = [ file for file in os.listdir(File_address) if file.endswith('.csv')]
    ax.set_title(title)
    for j in textfile:
        if ((j == 'Safety factor.csv') and ( output == 'Safety_factor')):
            d = pd.read_csv(File_address + '/' + j, skiprows=5)
            ax.plot(d.iloc[:,1])
            ax.set_ylabel(edgeAngle)
            ax.set_xlabel(output)  
#             ax.set_ylim(-1,15)
            
        elif ((j == 'Different energies.csv') and (output == 'Engy')):
            d = pd.read_csv(File_address + '/' + j, skiprows=5)
#             Total_contact_energy =  d.iloc[:,1].sum()
#             Elastic_strain_energy = d.iloc[:,3].max()
            ax.plot(d.iloc[:,1], 'r')
            ax.plot(d.iloc[:,3], 'b')
            ax.set_ylabel(edgeAngle)
            ax.set_xlabel(output)

        elif ((j == 'Maximum out-of-plane deformation.csv') and (output == 'Oop_deform')):
            d = pd.read_csv(File_address + '/' + j, skiprows=5)
    #             Max_oop_deformation = d.iloc[:,1].max()
            ax.plot(d.iloc[:,1])
            ax.set_ylabel(edgeAngle)
            ax.set_xlabel(output)
#             ax.set_ylim(-0.01,0.27)
            
        elif ((j == 'Maximum temperature of the edge.csv') and (output == 'Edge_temp')):
            d = pd.read_csv(File_address + '/' + j, skiprows=5)
#             Max_Edge_temp = d.iloc[:,1].max()
            ax.plot(d.iloc[:,1])
            ax.set_ylabel(edgeAngle)
            ax.set_xlabel(output)
            
        elif ((j == 'Average friction force.csv') and (output == 'Avr_frictForce')):
            d = pd.read_csv(File_address + '/' + j, skiprows=5)
#             Max_Ave_fric_force = d.iloc[:,1].mean()
            ax.plot(d.iloc[:,1])
            ax.set_ylabel(edgeAngle)
            ax.set_xlabel(output)
            
        elif ((j == 'Total net heat rate.csv') and (output == 'HeatRate')):
            d = pd.read_csv(File_address + '/' + j, skiprows=5)
#             Max_heat_rate = d.iloc[:,1].max()
            ax.plot(d.iloc[:,1])
            ax.set_ylabel(edgeAngle)
            ax.set_xlabel(output)
            
        elif ((j == 'Internal energy.csv') and (output == 'IntEngy')):
            d = pd.read_csv(File_address + '/' + j, skiprows=5)
#             Max_internal_energy = d.iloc[:,1].max()
            ax.plot(d.iloc[:,1])
            ax.set_ylabel(edgeAngle)
            ax.set_xlabel(output)
            
        elif ((j == 'Friction dissipation rate.csv') and (output == 'FricDissipRate')):
            d = pd.read_csv(File_address + '/' + j, skiprows=5)
#             Total_friction_dissipation = d.iloc[:,1].sum() 
            ax.plot(d.iloc[:,1])  
            ax.set_ylabel(edgeAngle)
            ax.set_xlabel(output)
```

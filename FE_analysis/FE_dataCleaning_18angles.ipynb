## We have simulation data for three different types of ceramic architectures

- 3 by 3 pannels with 4 angle variables
- 5 by 5 pannels with 6 angle variables
- 7 by 7 pannels with 8 angle variables

As the position of the tile borders are not fixed in defferent pattern, we define a data frame that includes three sets of angles, the first 4 angles for 3 by 3, the second 6 angles for 5 by 5 and the last eight angles for 7 by 7.

***The simulation data are saved in folders with different arrangments. We try to extract some of the data from their naming and pass the required information to the `output_calc` function to extract the required info from the `.csv` files in the passed address***  

```python
"""import packages you need"""
import numpy as np
import pandas as pd
import math
import os
import matplotlib.pyplot as plt 

from os import listdir
from os.path import isfile, join
```

```python
def output_calc(NumPiece, LengthRatio, File_address, Angles):
    
    if NumPiece == 9 :
        angle1_9 = Angles[0]
        angle2_9 = Angles[1]
        angle3_9 = Angles[2]
        angle4_9 = Angles[3]
        angle1_25 = angle2_25 = angle3_25 = angle4_25 = angle5_25 = angle6_25 = 0
        angle1_49 = angle2_49 = angle3_49 = angle4_49 = angle5_49 = angle6_49 = angle7_49 = angle8_49 = 0 
    
    elif NumPiece == 25 :
        angle1_25 = Angles[0]
        angle2_25 = Angles[1]
        angle3_25 = Angles[2]
        angle4_25 = Angles[3]
        angle5_25 = Angles[4]
        angle6_25 = Angles[5]
        angle1_9 = angle2_9 = angle3_9 = angle4_9 = 0
        angle1_49 = angle2_49 = angle3_49 = angle4_49 = angle5_49 = angle6_49 = angle7_49 = angle8_49 = 0 
    
    elif NumPiece == 49 :
        angle1_49 = Angles[0]
        angle2_49 = Angles[1]
        angle3_49 = Angles[2]
        angle4_49 = Angles[3]
        angle5_49 = Angles[4]
        angle6_49 = Angles[5]
        angle7_49 = Angles[6]
        angle8_49 = Angles[7]
        angle1_9 = angle2_9 = angle3_9 = angle4_9 = 0
        angle1_25 = angle2_25 = angle3_25 = angle4_25 = angle5_25 = angle6_25 = 0 
       
    textfile = [ file for file in os.listdir(File_address) if file.endswith('.csv')]
    for j in textfile:
        if (j == 'Safety factor.csv'):
            d = pd.read_csv(File_address + '/' + j, skiprows=5)
            Safety_factor = d.iloc[:,1].min()

        elif (j == 'Different energies.csv'):
            d = pd.read_csv(File_address + '/' + j, skiprows=5)
            Total_contact_energy =  d.iloc[:,1].sum()
            Elastic_strain_energy = d.iloc[:,3].max()

        elif (j == 'Maximum out-of-plane deformation.csv'):
            d = pd.read_csv(File_address + '/' + j, skiprows=5)
            Max_oop_deformation = d.iloc[:,1].max()
            
######### some of the simulation result missing 'Avr_reactForce'########### 
#             elif (j == 'Average reaction force.csv'):
#                 d = pd.read_csv(File_address + '/' + j, skiprows=5)
#                 Max_Ave_reaction_force = d.iloc[:,1].mean()
###########################################################################
        if (j == 'Maximum temperature of the edge.csv'):
            d = pd.read_csv(File_address + '/' + j, skiprows=5)
            Max_Edge_temp = d.iloc[:,1].max()

        elif (j == 'Average friction force.csv'):
            d = pd.read_csv(File_address + '/' + j, skiprows=5)
            Max_Ave_fric_force = d.iloc[:,1].mean()

        elif (j == 'Total net heat rate.csv'):
            d = pd.read_csv(File_address + '/' + j, skiprows=5)
            Max_heat_rate = d.iloc[:,1].max()

        elif (j == 'Internal energy.csv'):
            d = pd.read_csv(File_address + '/' + j, skiprows=5)
            Max_internal_energy = d.iloc[:,1].max()

        elif (j == 'Friction dissipation rate.csv'):
            d = pd.read_csv(File_address + '/' + j, skiprows=5)
            Total_friction_dissipation = d.iloc[:,1].sum() 
                
    DATA =  [NumPiece, LengthRatio,
                 angle1_9, angle2_9, angle3_9, angle4_9,
                 angle1_25, angle2_25, angle3_25, angle4_25, angle5_25, angle6_25,
                 angle1_49, angle2_49, angle3_49, angle4_49, angle5_49, angle6_49, angle7_49, angle8_49,
                 Safety_factor, Max_oop_deformation, Total_contact_energy,
                 Elastic_strain_energy, #Max_Ave_reaction_force,
                 Max_Edge_temp, Max_Ave_fric_force, Max_heat_rate, 
                 Max_internal_energy,Total_friction_dissipation
                 ]
#     print(Max_Edge_temp, File_address)    # for trouble shooting the files with different file names
    return DATA
```

```python
feature_names = ['Number_pieces', 'Length_ratio', 
             'angle1_9p','angle2_9','angle3_9p','angle4_9p',
             'angle1_25p','angle2_25p','angle3_25p','angle4_25p','angle5_25p','angle6_25p', 
             'angle1_49p','angle2_49p','angle3_49p','angle4_49p','angle5_49p','angle6_49p',
             'angle7_49p','angle8_49p',    ## some of the simulation result missing 'Avr_reactForce'
             'Safety_factor', 'Oop_deform', 'Tot_contactEngy',
             'Elast_strainEngy', 'Edge_temp', 'Avr_frictForce',
             'HeatRate', 'IntEngy', 'FricDissipRate'
            ]
```

## Constant Angles, LR = 1 

```python
mypath = '../FE_data/Constant_angle/3 Correct reaction/'
edgeAngle = [f for f in listdir(mypath) if f.endswith('deg')]

OUTPUTS = []

for i in range(len(edgeAngle)):
    angle = [int(s) for s in edgeAngle[i].split() if s.isdigit()][0]
    angles = [angle, angle , angle , angle]
    LengthRatio = 1
    NumP = 9 
    OUTPUTS.append(output_calc(NumP, LengthRatio, mypath + edgeAngle[i], angles))

Data = np.array([np.array(xi) for xi in OUTPUTS])

ConstAngle_3 = pd.DataFrame(Data, columns = feature_names)
# ConstAngle_3.to_csv(r'Data/FE_constAng3_29param.csv', header=True, index = False)

ConstAngle_3.iloc[:,:18]
```

# Variable Size
## Variable Size and Angle $3\times3$

```python
mypath = '../FE_data/Variable_size/3-3/'

OUTPUTS = []

edgeAngles = [f for f in listdir(mypath) if f.startswith('Case')]

for i in range(len(edgeAngles)):
    
    angles = [int(s) for s in edgeAngles[i].split() if s.isdigit()]
    lengthRatios = [lr for lr in listdir(mypath + '/' + edgeAngles[i]) if lr.startswith('LR=') ]
    
    for k in range(len(lengthRatios)):
        LengthRatio = float(lengthRatios[k][3:])
        FileAddress = mypath  + edgeAngles[i] + '/' + lengthRatios[k]
        NumP = 9
        OUTPUTS.append(output_calc(NumP, LengthRatio, FileAddress, angles))

Data = np.array([np.array(xi) for xi in OUTPUTS])

VariableSize_Angle3 = pd.DataFrame(Data, columns = feature_names)
VariableSize_Angle3.to_csv(r'Data/FE_Var_SizeAngle3_29param.csv', header=True, index = False)

VariableSize_Angle3.shape
```

## Variable Angle

```python
mypath = '../FE_data/Variable_angle/'

OUTPUTS = []
edgeAngles = [f for f in listdir(mypath) if f.startswith('Case ')]
for i in range(len(edgeAngles)):
    angles = [int(s) for s in edgeAngles[i].split() if s.isdigit()]

    blocks = [blk for blk in listdir(mypath + '/' + edgeAngles[i]) \
              if (blk.startswith('3') or blk.startswith('5') or blk.startswith('7'))]

    for k in range(len(blocks)):
        blk = blocks[k][0]
        FileAddress = (mypath  + edgeAngles[i]+str('/')+blocks[k])
        LengthRatio = 1
        NumP = int(blk) ** 2
        OUTPUTS.append(output_calc(NumP, LengthRatio, FileAddress, angles))

Data = np.array([np.array(xi) for xi in OUTPUTS])

VariableAngle = pd.DataFrame(Data, columns = feature_names)
VariableAngle.to_csv(r'Data/FE_VariableAngle_29param.csv', header=True, index = False)

VariableAngle.shape
```
